## Chapter-4 Class!

```python
class Toyota:
    def __init__(self, model, fuel_consumed, distance_traveled):
        self.model = model
        self.fuel_consumed = fuel_consumed
        self.distance_traveled = distance_traveled

    def mileage(self):
        if self.fuel_consumed == 0:
            return 0  
        return self.distance_traveled / self.fuel_consumed

car2 = Toyota("Supra", 35, 420)  
print(f"{car2.model} mileage: {car2.mileage():.2f} km/l")
```

```python
class Car:
    def __init__(self, name, model, year, color, price, engine_l, weight_kg, tyre_in):
        self.name = name
        self.model = model
        self.year = year
        self.color = color
        self.price = price
        self.engine_l = engine_l
        self.weight_kg = weight_kg
        self.tyre_in = tyre_in

    def display_info(self):
        print(f"Car Name: {self.name}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Color: {self.color}")
        print(f"Price: {self.price}")
        print(f"Engine: {self.engine_l} L")
        print(f"Weight: {self.weight_kg} kg")
        print(f"Tyre Size: {self.tyre_in} in")

    def mileage_info(self):
        base, city, highway = city_highway_kmpl(self.engine_l, self.weight_kg, self.tyre_in)
        print(f"Base mileage   : {base:.2f} km/L")
        print(f"City mileage   : {city:.2f} km/L")
        print(f"Highway mileage: {highway:.2f} km/L")


def dummy_mileage_kmpl(engine_l: float, weight_kg: float, tyre_in: float) -> float:
    mileage = 28 - 4 * engine_l - 0.006 * (weight_kg - 1000) - 0.7 * (tyre_in - 15)
    return max(5.0, mileage)


def city_highway_kmpl(engine_l: float, weight_kg: float, tyre_in: float) -> tuple[float, float, float]:
    base = dummy_mileage_kmpl(engine_l, weight_kg, tyre_in)
    city = 0.85 * base
    highway = 1.10 * base
    return base, city, highway


if __name__ == "__main__":
    my_car = Car("Toyota", "Supra MK4", 2002, "Red", 8500000, 3.0, 1615, 17)
    my_car.display_info()
    my_car.mileage_info()
```

## Create a class of animal(age,name,category) and make a subclass pet and inside sub class make a method vacintion

```python
class Animal:
    def __init__(self, age, name, category):
        self.age = age
        self.name = name
        self.category = category

    def vaccination(self):
        print(f"{self.name} has been vaccinated.")


class Pet(Animal):
    def pet_details(self):
        print("Pet Details:")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Category:", self.category)

        self.vaccination()


pet1 = Pet(3, "Bucky", "Dog")

pet1.pet_details()
```

## Design a Student Records Analyzer

```python
def build_valid_records(lines):
    records = {}

    for line in lines:
        try:
            roll, name, marks, section = [x.strip() for x in line.split(",")]
            roll, marks = int(roll), int(marks)

            if roll > 0 and name and 0 <= marks <= 100 and len(section) == 1 and section.isalpha():
                if roll not in records:
                    records[roll] = {"name": name, "marks": marks, "section": section}
        except:
            continue

    return records
```

```python
data = [
    "1, Shreyansh, 85, A",
    "2, Sandeep, 90, B",
    "3, Anushka, 78, A",
    "4, , 70, C",        # invalid (name empty)
    "5, Mahin, 105, B",    # invalid (marks >100)
    "6, Pranamya, 88, D",
    "7, Bhoomi, 92, E",
]

result = build_valid_records(data)

print(result)
```
