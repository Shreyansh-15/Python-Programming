```python
#Q1.
items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2
def top_k(items, k):
    items.sort(key=lambda x: x[1], reverse=True)
    return items[:k]
print(top_k(items, k))
```

```python
#Q2.
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word] =+ 1
    return d
```

```python
print(count_by_category(pairs))
```

```python
#Q3.
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) / 2
    return int(total - sum(nums))
```

```python
nums = [3, 0, 1]
print(missing_number(nums))
```

```python
# Q4.
def add_unique(x, seen=None):
    if seen is None:
        seen = set()
    if x not in seen:
        seen.add(x)
    return list(seen)
(print(add_unique(1)))
(print(add_unique(2)))
```
