## How would you do add operator/function overloading in Python
1.operator overloading

2.magical number

```python
class EncryptNumber:
    def __init__(self, num):
        self.num = str(num)

        self.digit_to_char = {
            '0': 'a', '1': 'b', '2': 'c', '3': 'd', '4': 'e',
            '5': 'f', '6': 'g', '7': 'h', '8': 'i', '9': 'j'
        }

        self.char_to_digit = {v: k for k, v in self.digit_to_char.items()}

    def encrypt(self):
        result = ""
        for d in self.num:
            result += self.digit_to_char[d]
        return result

    def decrypt(self, text):
        result = ""
        for ch in text:
            result += self.char_to_digit[ch]
        return int(result)

    def __add__(self, other):
        n1 = int(self.num)
        n2 = int(other.num)

        result = n1 + n2 + 1 
        return EncryptNumber(result)

    def __str__(self):
        return self.num

n1 = EncryptNumber(590012121)
n2 = EncryptNumber(0)

enc = n1.encrypt()
print("Encrypted:", enc)

res = n1 + n2
print("After +1:", res)
print("Decrypted:", res.num)
```

```python
class Sap:
    def __init__(self, num):
        self.num = str(num)

    def encript(self):
        result = ""
        for x in self.num:
            result += "abcdefghijklmnopqrstuvwxyz"[int(x)]
        return result
    
    def __add__(self, other):
        return Sap(int(self.num) + int(other.num) + 1)
    
    def __str__(self):
        return self.num
    
n = Sap(590012121)

print("Encrypted:", n.encript())
print("After +1:", n + Sap(0))
print("Decrypted:", n)
```

```python
class Student:
    def __init__(self, name, age, college):
        self.name = name
        self.age = age
        self.college = college

s1 = Student("Vish", 23, "UPES")
s2 = Student("Aman", 20, "DIT")

print(s1.name, s1.age, s1.college)
print(s2.name, s2.age, s2.college)
```

```python
from datetime import date

class Person:
    def __init__(self, name, birthdate):
        self.name = name
        self.birthdate = birthdate

    def convertToDays(self):
        today = date.today()
        return (today - self.birthdate).days

p1 = Person("Shreyansh", date(2002, 12, 15))
print(f"{p1.name}'s age in days is: {p1.convertToDays()} days")
```

```python
class Person:
    def __init__(self, name, age_years):
        self.name = name
        self.age_years = age_years

    def convertToDays(self):
        leap_days = self.age_years // 4   
        return self.age_years * 365 + leap_days

p1 = Person("Shreyansh", 23)
print(f"{p1.name}'s age in days is: {p1.convertToDays()} days")
```

##  Variable Scope ka Matlab
Scope ka matlab hai ek variable kahaan tak accessible hai — matlab program ke kis part mein usko use kar sakte ho.
Python mein scope decide karta hai ki variable "visible" hai ya nahi.
####  Types of Scope (LEGB Rule)
1. Local Scope
- Jo variable function ke andar banaya gaya hai.
- Sirf usi function ke andar use ho sakta hai.
2. Enclosed Scope
- Jab ek function ke andar dusra function ho (nested function).
- Inner function outer function ke variable ko access kar sakta hai.
3. Global Scope
- Jo variable function ke bahar banaya gaya hai.
- Pure program mein accessible hota hai.
4. Built-in Scope
- Python ke khud ke predefined names (jaise len, print, range).
- Ye hamesha available hote hain.
