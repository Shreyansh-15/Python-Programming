## Mutable and Immutable Data Types
Python objects are categorized based on whether their state can be modified after creation.

## :- Immutable Data Types
Immutable objects cannot be modified once created. If a change is required, Python creates a new
object in memory.

Common immutable types:
.int

.float

.bool

.str

.tuple

.frozenset


```python
s = "Hello, World!"
s = "World", "Python"
print(s)
```

## :-Mutable Data Types
Mutable objects allow in-place modification without changing the object identity.

-Common mutable types:

.list

.dict

.set

```python
numbers = [10, 20, 30]
numbers[0] = 99
print(numbers)
```
