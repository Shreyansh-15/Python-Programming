Q1. Class variable vs Instance variable

```python
class Student:
    school_name = "UPES"
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
s1 = Student("Aman", 20)
s2 = Student("Riya", 21)

print("School:", s1.school_name)
print("Student 1:", s1.name, s1.age)
print("Student 2:", s1.name, s2.age)
```

Q2. Instance Method and Class Method

```python
class Employee:
    company = "ABC Ltd"
    def __init__(self,name, salary):
        self.name = name
        self.salary = salary

    def show(self):
        print(self.name, self.salary, Employee.company)

    @classmethod
    def change(cls, name):
        cls.company = name 

e1 = Employee("Aman", 30000)  
e2 = Employee("Riya", 40000)

e1.show()
e2.show()

Employee.change("XYZ Ltd")
e1.show()
e2.show()
```

Q3. Static Method

```python
class MathHelper:
    @staticmethod
    def is_even(num):
        return num % 2 == 0
    
print(MathHelper.is_even(8))
print(MathHelper.is_even(11))
```

## Section B
Q4. Public and Private Data Members

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name 
        self.__balance = balance

    def deposit(self, amt):
        self.__balance += amt
    
    def withdraw(self, amt):
        if amt <= self.__balance:
            self.__balance -= amt
        else:
            print("Insufficient balance")
    
    def show_balance(self):
        print(self.__balance)

b = BankAccount("Aman", 1000)
b.deposit(500)
b.withdraw(200)
b.withdraw(2000)
b.show_balance()

```

## Section C
Q6. Magic Methods / Operator Overloading

```python
class Product:
    def __init__(self, name, price):
        self.name = name 
        self.price = price

    def __str__(self):
        return f"Product: {self.name}, Price: {self.price}"
    
    def __add__(self, other):
        return self.price + other.price
    
    def __eq__(self, other):
        return self.price == other.price
    
p1 = Product("Laptop", 50000)
p2 = Product("Phone", 30000)
p3 = Product("Tablet", 50000)

print(p1)
print(p1 + p2)  
print(p1 == p3)
print(p1 == p2)
```
